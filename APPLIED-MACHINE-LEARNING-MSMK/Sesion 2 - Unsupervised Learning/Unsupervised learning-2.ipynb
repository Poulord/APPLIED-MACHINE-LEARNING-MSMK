## Preprocess the data

The data must be preprocessed before training the network. If you inspect the first image in the training set, you will see that the pixel values fall in the range of 0 to 255:

```python
plt.figure()
plt.imshow(train_images[0])
plt.colorbar()
plt.grid(False)
plt.show()
```

Scale these values to a range of 0 to 1 before feeding them to the neural network model. To do so, divide the values by 255. It's important that the *training set* and the *testing set* be preprocessed in the same way:

```python
train_images = train_images / 255.0

test_images = test_images / 255.0
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.models import Model
from tensorflow.keras.layers import GlobalAveragePooling2D
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
def preprocess_gray_28_to_vgg(x, target_size=224):
    x = np.asarray(x)
    if x.ndim == 4 and x.shape[-1] == 1:
        x = x[..., 0]
    assert x.ndim == 3, "Se esperaba (N,28,28)"
    x3 = np.repeat(x[..., None], 3, axis=-1).astype('float32')
    if x3.max() <= 1.1:
        x3 *= 255.0
    x3 = tf.image.resize(x3, (target_size, target_size), antialias=True).numpy()
    x3 = preprocess_input(x3)
    return x3

def extract_features_vgg16(X, batch_size=128, layer='block5_pool', pooling='avg'):
    base_model = VGG16(weights='imagenet', include_top=False)
    output = base_model.get_layer(layer).output
    output = GlobalAveragePooling2D()(output)
    feature_model = Model(inputs=base_model.input, outputs=output)

    X_prep = preprocess_gray_28_to_vgg(X)
    features = []
    for i in range(0, len(X_prep), batch_size):
        batch = X_prep[i:i+batch_size]
        f = feature_model.predict(batch, verbose=0)
        features.append(f)
    return np.vstack(features)
```
